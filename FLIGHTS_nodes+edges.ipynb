# Upload Data

```python
df = pd.read_csv('/content/flights_cleaned.csv')
df
```

```python
df = df.reset_index()
df = df.drop(columns = ['lalala'])
df
```

# Create Nodes for flights

```python
name_to_id = {}
current_id = 0

for name in df['passengers']:
    if name not in name_to_id:
        name_to_id[name] = current_id
        current_id += 1

# Create the nodes DataFrame
nodes_df = pd.DataFrame({
    'Id': [name_to_id[name] for name in df['passengers']],
    'Name': df['passengers'],
    'Metadata': df['flight number ']
})

nodes_df
```

```python
nodes_df = nodes_df.groupby(['Name', 'Id'])['Metadata'].apply(list).reset_index()

nodes_df.to_csv(path + 'FLIGHTS_nodes.csv', index=False)
```

# Create Edges for Flights

```python
name_to_id = {}
current_id = 0

for name in df['passengers']:
    if name not in name_to_id:
        name_to_id[name] = current_id
        current_id += 1

# Create the nodes DataFrame
nodes_df = pd.DataFrame({
    'Id': [name_to_id[name] for name in df['passengers']],
    'Name': df['passengers'],
    'Metadata': df['flight number ']
})

nodes_df
```

```python
edges = []

grouped = nodes_df.groupby ('Metadata')

for _, group in grouped:
    for i in range(len(group)):
        for j in range(i+1, len(group)):
            edges.append((group.iloc[i]['Id'], group.iloc[j]['Id'], group.iloc[i]['Name'], group.iloc[j]['Name'], group.iloc[i]['Metadata'], group.iloc[j]['Metadata']))

edges_df = pd.DataFrame(edges, columns=['Source', 'Target', 'Source_name', 'Target_name', 'Metadata_1', 'Metadata_2'])

edges_df
```

```python
weighing_flights = edges_df.groupby(['Source', 'Target', 'Source_name', 'Target_name']).agg(lambda x: list(x))
weighing_flights
```

```python
weighing_flights['Weight'] = weighing_flights['Metadata_1'].apply(lambda x: len(x))
```

```python
weighing_flights
```

```python
weighing_flights = weighing_flights.reset_index()

weighing_flights.to_csv(path + 'final_edges.csv', index=False)
```
