# Upload Data

```python
flights_check = pd.read_excel('/content/flights_cleaned.xlsx')
flights_check
```

Find passengers that are alone and delete the related rows (they would not be part of a network)

```python
def is_lonely(passenger):
    return ',' not in passenger and '+' not in passenger

flights_check = flights_check[~flights_check['passengers'].apply(is_lonely)]

flights_check
```

Find duplicates and delete them

```python
# Identify duplicate flights with the same date
duplicate_flights_with_same_date = []

# Grouping by flight number and checking if any group has more than one entry
for flight_number, group in flights_check.groupby('flight number '):
    if len(group) > 1:
        # Checking if all dates within the group are the same
        if group['Date'].nunique() == 1:
            duplicate_flights_with_same_date.append(flight_number)

# Drop one of the duplicate rows for each flight number
for flight_number in duplicate_flights_with_same_date:
    flights_check.drop_duplicates(subset=['flight number ', 'Date'], keep='first', inplace=True)


flights_check
```

Find duplicates (same flight number, different date) and assign a proxy value to the flight number

```python
duplicate_flights = []

# Grouping by flight number and checking if any group has more than one entry
for flight_number, group in flights_check.groupby('flight number '):
    if len(group) > 1:
        duplicate_flights.append(flight_number)

duplicate_flights
```

```python
duplicate_flights = []

# Grouping by flight number and checking if any group has more than one entry
for flight_number, group in flights_check.groupby('flight number '):
    if len(group) > 1:
        duplicate_flights.append(flight_number)

# Iterate over each flight number with duplicates
for flight_number in duplicate_flights:
    # Get the groups of rows with the same flight number
    groups = flights_check.groupby('flight number ').get_group(flight_number)

    # Iterate over each group and update the flight number
    for i, (_, group) in enumerate(groups.iterrows()):
        new_flight_number = f"{flight_number}{chr(97 + i)}"
        flights_check.loc[group.name, 'flight number '] = new_flight_number

# Print the updated dataframe
flights_check
```

Create a new dataframe in which passengers do not belong to the same row

```python
new_df = pd.DataFrame(columns=flights_check.columns)

for index, row in flights_check.iterrows():
    passenger_list = [passenger.strip() for passenger in row['passengers'].split(',')]

    if len(passenger_list) == 1:
        flights_check = flights_check.drop(index)

    for passenger in passenger_list:
        new_row = row.copy()
        new_row['passengers'] = passenger
        new_df = new_df.append(new_row, ignore_index=True)
```

```python
new_df
```

Clean data with regular expressions

```python
import re
```

```python
pattern = r'(BABY)|(\b\w+\b\?)|(\?+)|(\d)|(\b\w+\b\s\?)|(\b\w+\b\s\b\w+\b\?)|(.+?\?)|(AND \d PAX)|(\+ \d PAX)|((\w+(?:\s+\w+)*\s*\+\s*\d+(?:\s+\w+)*)+)'

filtered_df = new_df[~new_df['passengers'].str.match(pattern)]
filtered_df
```

```python
filtered_df.to_csv(path + 'flights_cleaned.csv', index=True)
```
