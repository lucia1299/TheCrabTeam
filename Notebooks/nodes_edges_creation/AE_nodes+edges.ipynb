# Create Edges for AE

```python
edges = []
```

```python
grouped = df.groupby('Associated Entity')
```

```python
for _, group in grouped:
    for i in range(len(group)):
        for j in range(i+1, len(group)):
            edges.append((group.iloc[i]['index'], group.iloc[j]['index'], group.iloc[i]['Name'], group.iloc[j]['Name'], group.iloc[i]['Associated Entity'], group.iloc[j]['Associated Entity']))
```

```python
edges_df = pd.DataFrame(edges, columns=['Source', 'Target', 'Source_Name', 'Target_Name', 'Label1', 'Label2' ])

edges_df
```

```python
grouped = edges_df.groupby(['Source', 'Target', 'Source_Name', 'Target_Name']).agg(lambda x: list(x))
grouped = grouped.reset_index()

grouped
```

```python
grouped['Weight'] = grouped.apply(lambda row: len(row['Label1']), axis=1)
grouped
```

```python
edges_df.to_csv(path + 'edges_AE.csv', index=False)
```
