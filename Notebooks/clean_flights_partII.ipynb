```python
import pandas as pd

my_flights = pd.read_excel("/flights_cleaned.xlsx")

# Create an empty DataFrame to store the new rows
new_df = pd.DataFrame(columns=my_flights.columns)

# Iterate over each row in my_flights
for index, row in my_flights.iterrows():
    passenger_list = row['passengers'].split(',')

    # Iterate over each passenger in passenger_list
    for passenger in passenger_list:
        # Create a new row for each passenger
        new_row = row.copy()
        new_row['passengers'] = passenger  # Replace 'passengers' value with the current passenger
        new_df = new_df.append(new_row, ignore_index=True)

# Now new_df contains the new rows with individual passengers
```
