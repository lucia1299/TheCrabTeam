```python
import pandas as pd
```

```python
path = 'drive/MyDrive/'
```

```python
df1 = pd.read_csv(path + 'epstein_black_book.csv', usecols=['Name','Type', 'City', 'Associated Entity', 'Associated Terms'])
df2 = pd.read_csv(path + 'epstein_black_book2.csv', usecols=['Name','Type', 'City', 'Associated Entity', 'Associated Terms'])
```

```python
df1
```

```python
df2
```

```python
merged_df = pd.merge(df1, df2, on=['Name','Type', 'City', 'Associated Entity', 'Associated Terms'], how='outer')
```

```python
merged_df
```

```python
df_person = merged_df[merged_df['Type'] == 'Person']
```

```python
df_person = df_person.reset_index()
```

```python
df_person
```

```python
df_city = df_person.dropna(subset=["City"]).reset_index()
```

```python
df_city
```

```python
df_ae = df_person.dropna(subset=["Associated Entity"]).reset_index()
```

```python
df_ae
```

```python
merged_df.to_csv(path + 'epstein_merged.csv', index=True)
```

```python
df_city.to_csv(path + 'black_books_city.csv', index=True)
```

```python
df_ae.to_csv(path + 'black_books_ae.csv', index=True)
```
